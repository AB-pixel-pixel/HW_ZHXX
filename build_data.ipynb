```python
from pydub import AudioSegment

# 加载音频文件
audio = AudioSegment.from_mp3(r"TTS/video_voice.mp3")


def save_sliced_audio(audio, start_second, end_second, save_counter):
    # 切片音频
    sliced_audio = audio[start_second * 1000:end_second * 1000]  # 设置切片时间 (毫秒)

    # 保存切片后的音频
    sliced_audio.export(f"TTS/train_data/output{save_counter}.mp3", format="mp3")


def get_file(file_path):
    labeled_seconds = [0]
    # 打开文件
    with open(file_path, 'r') as file:
        # 逐行读取
        for line in file:
            # 处理每一行数据
            print(line, end='')  # end='' 用于避免打印空行
            _time = line.split('|')[1]
            minute, second = _time.split('.')
            total_second = int(minute) * 60 + int(second)
            labeled_seconds.append(total_second)
    return labeled_seconds


labeled_second = get_file('TTS/label_list.txt')
for counter in range(len(labeled_second)):
    if counter != 0:
        save_sliced_audio(audio, labeled_second[counter - 1], labeled_second[counter], counter)
```

```python
from pydub import AudioSegment
import os

# 设置目标时长 (毫秒)
target_duration = 35 * 1000  

# 训练数据文件夹路径
train_data_dir = "train_data_pre_pad"

# 遍历所有 MP3 文件
for filename in os.listdir(train_data_dir):
    if filename.endswith(".mp3"):
        # 加载音频文件
        audio_path = os.path.join(train_data_dir, filename)
        audio = AudioSegment.from_mp3(audio_path)

        # 获取音频时长
        audio_duration = len(audio)

        # 填充或截断音频
        if audio_duration < target_duration:
            # 填充: 在音频末尾重复自身，直到达到目标时长
            silence = AudioSegment.silent(duration=target_duration - audio_duration)
            audio = audio + silence
        elif audio_duration > target_duration:
            # 截断: 保留音频开头部分，截取到目标时长
            audio = audio[:target_duration]

        # 保存处理后的音频
        audio.export(os.path.join("train_data", filename), format="mp3")

print("所有音频文件已统一到 35 秒！")
```

```python
import librosa
import os
import numpy as np

def find_min_max_db(folder_path):
  """查找文件夹中所有音频文件的最小和最大 dB 值。

  Args:
    folder_path: 包含音频文件的文件夹路径。

  Returns:
    一个元组，包含最小 dB 值和最大 dB 值。
  """
  min_db = float("inf")
  max_db = float("-inf")

  for filename in os.listdir(folder_path):
    if filename.endswith(".mp3"):
      file_path = os.path.join(folder_path, filename)
      audio, sr = librosa.load(file_path)
      
      # 计算音频的 dB 值
      db = librosa.amplitude_to_db(np.abs(librosa.stft(audio)), ref=0)
      
      # 更新最小和最大 dB 值
      min_db = min(min_db, np.min(db))
      max_db = max(max_db, np.max(db))

  return min_db, max_db

# 文件夹路径
folder_path = "train_data_pre_pad" 

# 获取最小和最大 dB 值
min_db, max_db = find_min_max_db(folder_path)

# 打印结果
print(f"最小 dB 值: {min_db:.2f} dB")
print(f"最大 dB 值: {max_db:.2f} dB")
```
